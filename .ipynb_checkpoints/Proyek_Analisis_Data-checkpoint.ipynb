## Exploratory Data Analysis (EDA)

### Explore ...

```python
data_hr.describe(include="all")
```

```python
# Analisis pola penggunaan sepeda berdasarkan musim
season = data_day.groupby("season").agg({
    'cnt': ['mean', 'sum'],  # Rata-rata dan total jumlah penyewaan
    'casual': 'mean',        # Rata-rata pengguna kasual
    'registered': 'mean'     # Rata-rata pengguna terdaftar
}).reset_index()

print(season)
```

```python
#jumlah penyewaan berdasarkan musim
result = data_hr.groupby("season")[["cnt", "casual", "registered"]].mean().reset_index()

result_sorted = result.sort_values(by=["cnt", "casual", "registered"], ascending=False)

print(result_sorted)
```

```python
#Menghitung rata-rata suhu untuk setiap hari dalam seminggu
data_hr.groupby(by="weekday").temp.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
data_day.describe(include="all")
```

```python
# Menggabungkan data berdasarkan kolom dteday
data_combined = pd.merge(data_day, data_hr, on="dteday", suffixes=('_day', '_hour'))

# Cek penggabungan
print(data_combined.head())
```

```python
# Mendapatkan kolom dari kedua dataset
day_columns = set(data_day.columns)
hour_columns = set(data_hr.columns)

# Kolom yang hanya ada di day.csv
only_in_day = day_columns - hour_columns

# Kolom yang hanya ada di hour.csv
only_in_hour = hour_columns - day_columns

print("Kolom yang hanya ada di day.csv:", only_in_day)
print("Kolom yang hanya ada di hour.csv:", only_in_hour)
```

```python
# Melihat hasil analisis rata-rata penyewaan per jam
hourly_pattern = data_hr.groupby("hr")[["cnt", "casual", "registered"]].mean().reset_index()
print(hourly_pattern)
```

```python
data_numerik = data_hr.select_dtypes(include=["float64", "int64"])

# Menghitung matriks korelasi
korelasi = data_numerik.corr()

# Korelasi dengan kolom target `cnt`
korelasi_cnt = korelasi["cnt"].sort_values(ascending=False)
print(korelasi_cnt)
```

**Insight:**
- rata-rata jumlah penyewaan sepeda (cnt), baik untuk pengguna casual maupun registered, mencapai nilai tertinggi pada musim 3 dibandingkan dengan musim lainnya
-  melihat apakah ada perbedaan suhu yang signifikan antara hari-hari tertentu dalam seminggu

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pola penggunaan sepeda bedasarkan musim?



```python
season.columns = ['season', 'cnt_mean', 'cnt_sum', 'casual_mean', 'registered_mean']

# Plot rata-rata penyewaan
plt.figure(figsize=(6, 7))
sns.barplot(x='season', y='cnt_mean', data=season, palette='viridis', hue='season', dodge=False)
plt.title('Rata-rata Penyewaan Sepeda per Musim', fontsize=14)
plt.xlabel('Musim', fontsize=12)
plt.ylabel('Rata-rata Penyewaan', fontsize=12)
plt.xticks(ticks=[0, 1, 2, 3], labels=['Winter', 'Spring', 'Summer', 'Fall'])
plt.legend([], [], frameon=False)  
plt.show()
```

### Pertanyaan 2: Faktor apa yang paling memengaruhi jumlah pengguna sepeda?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="hr", y="cnt", data=hourly_pattern, label="Total Penyewaan", color="blue")
sns.lineplot(x="hr", y="casual", data=hourly_pattern, label="Casual", color="green")
sns.lineplot(x="hr", y="registered", data=hourly_pattern, label="Registered", color="orange")

# Menambahkan detail plot
plt.title("Pola Rata-rata Penyewaan Sepeda per Jam", fontsize=14)
plt.xlabel("Jam", fontsize=12)
plt.ylabel("Rata-rata Penyewaan", fontsize=12)
plt.xticks(range(0, 24))
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
korelasi_cnt.drop("cnt").sort_values(ascending=True).plot(kind="barh", color="skyblue")
plt.title("Korelasi Variabel dengan `cnt`", fontsize=14)
plt.xlabel("Korelasi", fontsize=12)
plt.ylabel("Variabel", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

**Insight:**
- Strategi promosi seperti diskon untuk pengguna kasual di musim dingin atau pengaturan ulang stok sepeda berdasarkan musim dapat membantu mengoptimalkan penggunaan sepeda.
-  Fokus pada pengguna terdaftar dan jam sibuk dapat meningkatkan layanan, misalnya menyediakan lebih banyak sepeda di stasiun tertentu selama waktu sibuk.

## Conclusion

- 1. strategi musiman diperlukan untuk memaksimalkan penggunaan sepeda. Promosi khusus, seperti diskon atau acara komunitas, dapat diterapkan selama musim dengan penggunaan rendah (musim dingin)
- 2. meningkatkan layanan untuk pengguna terdaftar (registered), seperti pemeliharaan sepeda dan promosi pada jam sibuk (pagi dan sore), dapat mengoptimalkan jumlah penyewaan(cnt)
